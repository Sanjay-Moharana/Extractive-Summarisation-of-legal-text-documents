```python
import os
import json
import gc
```

```python
# Reading the DF Corpus
with open('TFCorpus.json', 'r') as fp:
    readTF = json.load(fp)
fp.close()
gc.collect()
```

```python
# Reading the IDF Corpus
with open('IDFCorpus.json', 'r') as fp:
    readIDF = json.load(fp)
fp.close()
gc.collect()
```

```python
# Reading the Unique word Corpus
with open('UniqueBowOfDocForIDFCorpus.json', 'r') as fp:
    readUniqueWord = json.load(fp)
fp.close()
gc.collect()
```

```python
# Reading the IDF Corpus
with open('ParagraphWordCorpus.json', 'r') as fp:
    readParaWord = json.load(fp)
fp.close()
gc.collect()
```

```python
# Reading the IDF Corpus
with open('ParagraphCorpus.json', 'r') as fp:
    readPara = json.load(fp)
fp.close()
gc.collect()
```

```python
#reading the document
fileName = input('Please insert the name of the document : ')
```

```python
uniqueWords = readUniqueWord[fileName] # unique words of the document
paragraphsWithWords = readParaWord[fileName] # list of paragraph with words
paragraphsWithSents = readPara[fileName][0] # list of paragraph with sentences
paragraph = []
paragraphVector = []
dictTF = readIDF[fileName] # TF of words in the document
dictIDF = readTF[fileName] # IDF of words in the document

# computing the document vectors
for para in readParaWord[fileName]:
    for word in readUniqueWord[fileName]:
        flag = 0
        for sentence in para:
            if(word in sentence):
                tfIdfScore = dictTF[word] * dictIDF[word]
                paragraph.append(tfIdfScore)
                flag = 1
                break
        if(flag == 0):
             paragraph.append(0)
    paragraphVector.append(paragraph)
    paragraph = []
```

```python
print (len(paragraphVector))
```

```python
# Allows charts to appear in the notebook
%matplotlib inline
import matplotlib.pyplot as plt

import networkx as nx
import numpy as np
```

```python
#creating a empty graph with no node and no edges
graph = nx.Graph()

#creating nodes from each sentenses
graph.add_nodes_from([index for index in range(len(paragraphVector))])
```

```python
#function to compute cosine similarity of the two tf-idf vector
def computeCosineSimilarity(vector1, vector2):
    return np.dot(vector1, vector2)/(np.linalg.norm(vector1) * np.linalg.norm(vector2))
```

```python
#computing cosine similarity
for i in range(len(paragraphVector)-1):
    j = i + 1
    while(j < len(paragraphVector)):
        graph.add_edge(i, j, weight=computeCosineSimilarity(paragraphVector[i], paragraphVector[j])) #(u, v) edge
        j += 1
```

```python
#ranking the page
pageRank = nx.pagerank(graph)
# print(pageRank)
#fetching the heghest K page rank score sentence indices
sentenseIndex = sorted(pageRank, key=pageRank.get, reverse=True)[:10]

#sorting the pageranks in ascending order to retain the flow and important sentences as the appear in the document
sentenseIndex.sort()

#printing the summary
print("Top Paragraphs\n-------------------------")
for index in sentenseIndex:
    print (*paragraphsWithSents[index])
```
